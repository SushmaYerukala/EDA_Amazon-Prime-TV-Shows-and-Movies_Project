## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# ----- Handling Missing Values -----

# Fill missing age certification
titles['age_certification'] = titles['age_certification'].fillna('Unknown')

# Fill missing genres
# Based on `titles.isnull().sum()`, `genres` has no missing values before this step.
# If there were NaN, this would fill them with the string '[]'.
titles['genres'] = titles['genres'].fillna('[]')

# Fill seasons only for movies
titles['seasons'] = titles['seasons'].fillna(0)

# Drop rows where IMDb score is missing
titles = titles.dropna(subset=['imdb_score'])

# ----- Convert genres from string list to actual Python list -----
import ast

def convert_genres_to_list(item):
    if isinstance(item, str):
        try:
            # Attempt to evaluate the string as a Python literal (e.g., '["action"]')
            return ast.literal_eval(item)
        except (ValueError, SyntaxError):
            # If it's a string but not a valid list literal (e.g., just 'Action'),
            # wrap it in a list.
            return [item]
    elif isinstance(item, list):
        # If it's already a list, return it as is.
        return item
    else:
        # Handle other unexpected types, converting to an empty list
        return []

titles['genres'] = titles['genres'].apply(convert_genres_to_list)


# ----- Handling Outliers (Optional) -----
# Remove impossible IMDb ratings
titles = titles[(titles['imdb_score'] >= 0) & (titles['imdb_score'] <= 10)]

# Reset index after cleaning
titles = titles.reset_index(drop=True)
````

#### Chart - 12

````python
# Chart - 12 visualization code

# Define score categories
def get_score_category(score):
    if score < 4:
        return 'Poor'
    elif 4 <= score < 6:
        return 'Average'
    elif 6 <= score < 8:
        return 'Good'
    else:
        return 'Excellent'

# Apply the function to create a new 'imdb_score_category' column
titles['imdb_score_category'] = titles['imdb_score'].apply(get_score_category)

# Count the occurrences of each category
score_category_counts = titles['imdb_score_category'].value_counts()

# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(score_category_counts, labels=score_category_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis'))
plt.title('Distribution of Content by IMDb Score Category on Amazon Prime Video')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````

##### 1. Why did you pick the specific chart?

A pie chart is excellent for visualizing the proportion of each part to the whole. By categorizing IMDb scores, it effectively shows the percentage of content falling into 'Poor', 'Average', 'Good', or 'Excellent' quality ranges.

##### 2. What is/are the insight(s) found from the chart?

This chart reveals the overall quality distribution of content. We can see if Amazon Prime Video's library predominantly consists of 'Average' content, has a strong showing of 'Good' and 'Excellent' titles, or if a significant portion falls into 'Poor' quality.
